```python
import sqlite3
import pandas as pd
import savantscraper

# Example for two years and two teams
savantscraper.database_import('baseball_savant_all',
                              (2018, 2018))
```

```python
# Connect to the database
conn = sqlite3.connect('baseball_savant_all.db')

# Query the database and load into a pandas dataframe

df = pd.read_sql_query(
    "select pitch_type, release_speed, release_pos_x, release_pos_z, pfx_x, pfx_z, stand, p_throws, "
    "plate_x, plate_z, release_pos_y, events from statcast "
    "where events in ('single', 'double', 'triple', 'home_run', 'field_out');", conn)

# Close connection when finished
conn.close()

df
```
